In this space Exploratory Data Analysis will be performed.
We start by importing all libraries that will be used.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

We need to load in our .csv file.

```python
df = pd.read_csv('principal_components.csv')
```

Once we have this it is important t osee basic information on the data, as well as the summary statistics, and check if there are any missing values.

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

Histograms will be made of the principal components to help display visually how much variance is captured in each component.

```python
df.hist(figsize=(8, 6))
plt.suptitle('Histograms of Principal Components')
plt.show()
```

Box plots of the principal components will display the spread of data so we can see the outliers compared to the main data distribution.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df)
plt.title('Boxplot of Principal Components')
plt.show()
```

A pairwise visual will be used to plot both components agaisnt eash other, this will help to reveal possible correlations and patterns.

```python
sns.pairplot(df)
plt.show()
```

A correlation matrix wil show the strenths and relationships of the principal components and how strong their influence maybe.

```python
# Compute and visualize the correlation matrix
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix of Principal Components')
plt.show()
```
