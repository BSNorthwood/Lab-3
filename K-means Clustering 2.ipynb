The first thing we will always do is top import all libraries we will be using.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

Next we load in our csv file

```python
df = pd.read_csv('simulated_health_wellness_data.csv')
```

We must designate the feature columns

```python
X = df[['Stress_Level', 'Exercise_Time_Min']]
```

For a K-means clustering we will standardize the data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

We set the amount of clusters and the random state to ensure reproducability

```python
kmeans = KMeans(n_clusters=3, random_state=42)  # Choose the number of clusters
df['cluster'] = kmeans.fit_predict(X_scaled)
```

Now that we have initiated the k-means clustering we have the code to show the plot

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['cluster'], cmap='viridis')
plt.xlabel('Stress')
plt.ylabel('Exercise_Time_Min')
plt.title('K-Means Clustering')
plt.show()
```
