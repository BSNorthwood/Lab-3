We will be running the Hierarchical clustering one the PCA set.
First we import all needed libraries.

```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

We need to load in our .csv file

```python
principal_df = pd.read_csv('principal_components.csv')
```

We are using linkage to condense the data points that merge. We use ward to minimize variance.

```python
linkage_matrix = linkage(principal_df, method='ward')
```

Now that we have set all the background work we make the visual and display it.

```python
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix, labels=principal_df.index, leaf_rotation=90, leaf_font_size=10)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
